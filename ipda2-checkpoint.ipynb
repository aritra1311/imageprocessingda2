## Installing required packages

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('7.jpg', flags=cv.IMREAD_COLOR)
```

```python
plt.imshow(img)
```

## Sharpening the images with a kernel of 3*3 dimension

```python
kernel = np.array([[0, -1, 0],
                   [-1, 100,-1],
                   [0, -1, 0]])
img_sharp = cv.filter2D(src=img, ddepth=-1, kernel=kernel)
plt.imshow(img)
## We can see here that the image has been sharpened and the edges are better than it's orginal one
```

```python
##Avergae smoothning
##blurring with a 7*7 kernel

img = cv.imread('1.jpg')

blur = cv.blur(img,(7,7))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## Gausian blur
img = cv.imread('1.jpg')

blur = cv.GaussianBlur(img,(7,7),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## Median Blur
img = cv.imread('1.jpg')
median = cv.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
